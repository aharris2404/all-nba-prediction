```python
import pandas as pd
import os
```

```python
files = os.listdir(os.path.join(os.getcwd(), '../data/chunks/'))

files.sort()

print(files)
```

```python
mod = '../data/chunks/'
chunks_dir = os.path.join(os.getcwd(), mod)
files = os.listdir(chunks_dir)
files.sort()
print(files)
```

```python
final_df = pd.DataFrame()

for i, filename in enumerate(files):
    print(filename)
    df = pd.read_csv(os.path.join(chunks_dir, filename), index_col=0)
    print(df.head())
    if i == 0:
        final_df = df
    else:
        final_df = pd.concat([final_df, df])
```
